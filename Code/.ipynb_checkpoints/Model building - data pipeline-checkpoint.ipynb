```python
#graphs
def graph_nationality(artists):
    plt.figure(figsize=(18,12))
    nationality=sns.barplot(data=artists.groupby("nationality").sum().reset_index().sort_values("paintings",ascending=False),x="nationality",y="paintings")
    plt.xlabel("Nationality")
    plt.ylabel("Number of paintings")
    plt.xticks(rotation='vertical')
    plt.title("Most proficient nationalities over the top 50 artists",fontsize=15)
    plt.show()
    return nationality

def graph_painters(artists):
    plt.figure(figsize=(18,12))
    plt.xticks(rotation='vertical')
    painters=sns.barplot(data=artists[["name","paintings"]].sort_values("paintings",ascending=False),x="name",y="paintings")
    plt.xlabel("Names of painters")
    plt.ylabel("Number of paintings")
    plt.title("Ranking of painters from their number of paintings",fontsize=15)
    plt.show()
    return painters

def graph_genre1(artists):
    plt.figure(figsize=(18,8))
    genre1=sns.barplot(data=pd.DataFrame(artists.genre.value_counts()).reset_index(),x="index",y="genre")
    plt.xlabel("Genre")
    plt.ylabel("Occurrence of genre")
    plt.xticks(rotation='vertical')
    plt.title("Genre representation over the top 50 artists",fontsize=15)
    plt.show()
    return genre1

def genre_cleaning(artists):
    dct_genre={"Abstract Expressionism":"Expressionism","Surrealism,Impressionism":"Surrealism","Early Renaissance":"Renaissance","Expressionism,Abstractionism" :"Expressionism","Expressionism,Abstractionism,Surrealism":"Expressionism","High Renaissance":"Renaissance","High Renaissance,Mannerism":"Renaissance","Impressionism,Post-Impressionism":"Impressionism","Northern Renaissance":"Renaissance","Primitivism,Surrealism":"Primitivism","Proto Renaissance":"Renaissance","Realism,Impressionism":"Realism","Social Realism,Muralism":"Realism"," Symbolism,Art Nouveau":"Symbolism","Symbolism,Expressionism":"Symbolism","Symbolism,Post-Impressionism":"Symbolism","Symbolism,Art Nouveau":"Symbolism"}
    artists.genre=artists.genre.map(dct_genre).fillna(artists.genre)
    return artists

def graph_genre2(artists):
    plt.figure(figsize=(18,8))
    genre2=sns.barplot(data=pd.DataFrame(artists.genre.value_counts()).reset_index(),x="index",y="genre")
    plt.xlabel("Genre")
    plt.ylabel("Occurrence of genre")
    plt.xticks(rotation='vertical')
    plt.title("Genre representation over the top 50 artists",fontsize=15)
    plt.show()
    return genre2

# Images

def resize_images():
    for i in os.listdir():
        for j in os.listdir(i):
            load_img_rz = np.array(Image.open(r'C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100/'+str(i)+"/"+str(j)).resize((100,100)))
            Image.fromarray(load_img_rz).save(r'C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100/'+str(i)+'/'+str(j))
            
def load_image( infilename ) :
    img = Image.open( infilename )
    img.load()
    data = np.asarray( img, dtype="int32")
    return data

#CREATING X

def create_X():
    X_list1=[]
    X_list2=[]
    index1=[]
    for i in os.listdir("C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100_top_5/"):
        for j in os.listdir(i):
            result = load_image(r'C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100_top_5/'+(i)+'/'+(j))
            if result.shape != (100, 100,3):
                index1.append(j)
                X_list1.extend([result])

            else:
                X_list2.extend([result])
    return X_list2

def index1_def():
    X_list1=[]
    X_list2=[]
    index1=[]
    for i in os.listdir("C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100_top_5/"):
        for j in os.listdir(i):
            result = load_image(r'C:/Users/Camille/Documents/Ecole/Ironhack/Week 9/Final Project/images_resized_100_top_5/'+(i)+'/'+(j))
            if result.shape != (100, 100,3):
                index1.append(j)
                X_list1.extend([result])

            else:
                X_list2.extend([result])
    return index1

#CREATING Y
def create_y(index1):
    dct_images={j:i for i in os.listdir() if '.' not in i for j in os.listdir(i) if j.endswith("jpg") if j not in index1 }
    class_names=['Impressionism', 'Renaissance', 'Post-Impressionism', 'Symbolism', 'Baroque']
    dct_class={"Impressionism":0,"Renaissance":1,"Post-Impressionism":2,"Symbolism":3,"Baroque":4}
    y_df=pd.DataFrame(list(dct_images.values()),columns=["Genre"])
    y_df.Genre=y_df.Genre.map(dct_class)
    y=np.array(list(y_df.Genre))
    return y

def original_genre():
    plt.figure(figsize=(10,10))
    for i in range(9):
        original=plt.subplot(3,3,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(X_train[i], cmap=plt.cm.binary)
        plt.xlabel(genre[y_train[i]])
    plt.show()
    return original

# Image transformation
def image_transformation1():
    datagen1 = ImageDataGenerator(rotation_range=90)
    # configure batch size and retrieve one batch of images
    plt.figure(figsize=(10,10))
    plt.grid(False)
    for X_batch1, y_batch1 in datagen1.flow(X_train, y_train, batch_size=500):
        for i in range(9):
            transformed1=plt.subplot(3,3,i+1)
            plt.xticks([])
            plt.yticks([])
            #plt.subplot(330 + 1 + i)
            plt.imshow(X_batch1[i])
            plt.xlabel(genre[y_batch1[i]])
    # show the plot
        plt.show()
        break
    return transformed1,X_batch1, y_batch1 

def image_transformation2():
    datagen2 = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)
    # configure batch size and retrieve one batch of images
    plt.figure(figsize=(10,10))
    plt.grid(False)
    for X_batch2, y_batch2 in datagen2.flow(X_train, y_train, batch_size=255):
        for i in range(9):
            transformed2=plt.subplot(3,3,i+1)
            plt.xticks([])
            plt.yticks([])
            plt.imshow(X_batch2[i])
            plt.xlabel(genre[y_batch2[i]])
    # show the plot
        plt.show()
        break
    return  transformed2,X_batch2, y_batch2

def image_transformation3():
    datagen3 = ImageDataGenerator(zoom_range=[0.5,1.0])
    plt.figure(figsize=(10,10))
    plt.grid(False)
    for X_batch3, y_batch3 in datagen3.flow(X_train, y_train, batch_size=255):
        for i in range(9):
            transformed3=plt.subplot(3,3,i+1)
            plt.xticks([])
            plt.yticks([])
            plt.imshow(X_batch3[i])
            plt.xlabel(genre[y_batch3[i]])
    # show the plot
        plt.show()
        break
    return transformed3,X_batch3, y_batch3 

def X_train_transformed(X_train,X_batch1,X_batch2,X_batch3):
    X_train2=np.append(X_train, X_batch1, axis=0)
    X_train2=np.append(X_train2, X_batch2, axis=0)
    X_train2=np.append(X_train2, X_batch3, axis=0)
    return X_train2

def y_train_transformed(y_train,y_batch1,y_batch2,y_batch3):
    y_train2=np.append(y_train, y_batch1, axis=0)
    y_train2=np.append(y_train2, y_batch2, axis=0)
    y_train2=np.append(y_train2, y_batch3, axis=0)
    return y_train2

#Creating model

def model(X_train,X_test,y_train,y_test):
    #model creating
    model = keras.Sequential([
    keras.layers.Flatten(input_shape=(100, 100,3)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(5)])
    #model compile
    model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=10, validation_split=0.2,shuffle=True,use_multiprocessing=True,workers=5) #adding validation
    return model

def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(genre[predicted_label],
                                100*np.max(predictions_array),
                                genre[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')

#Results

def plot_results(predictions):
    # Plot the first X test images, their predicted labels, and the true labels.
    # Color correct predictions in blue and incorrect predictions in red.
    num_rows = 5
    num_cols = 3
    num_images = num_rows*num_cols
    plt.figure(figsize=(2*2*num_cols, 2*num_rows))
    for i in range(num_images):
        results=plt.subplot(num_rows, 2*num_cols, 2*i+1)
        plot_image(i, predictions[i], y_test, X_test)
        #plt.subplot(num_rows, 2*num_cols, 2*i+2)
        #plot_value_array(i, predictions[i], y_test)
    plt.tight_layout()
    plt.show()
    return results

def confusion_matrix_plot(X_test,y_test):
    preds=model.predict(X_test)
    array=confusion_matrix(y_test, preds.argmax(axis=1),labels=[0,1,2,3,4])
    df_cm = pd.DataFrame(array, range(5), range(5))
    df_cm=df_cm.rename(index={0: "Impressionism", 1: "Renaissance", 2: "Post-Impressionism",3:"Symbolism",4:"Baroque"})
    df_cm=df_cm.rename(columns={0: "Impressionism", 1: "Renaissance", 2: "Post-Impressionism",3:"Symbolism",4:"Baroque"})
    plt.figure(figsize=(10,7)) # for label size
    plt.title("Confusion matrix"+"\n")
    confusion=sns.heatmap(df_cm, annot=True, annot_kws={"size": 12},cmap='Oranges',fmt='g') # font size
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    return confusion

def transform_url(url):
    # Predict from web 
    web_image = imageio.imread(url)
    web_image = cv2.resize(web_image, dsize=(100,100) )
    #web_image = image.img_to_array(web_image)
    web_image = web_image/255.0
    web_image = np.expand_dims(web_image, axis=0)
    return web_image

def predict_url(web_image):
    prediction = model.predict(web_image)
    prediction_probability = np.amax(prediction)
    prediction_idx = np.argmax(prediction)

    print("Predicted genre =", genre[prediction_idx])
    print("Prediction probability =", prediction_probability*100, "%")

    pred=plt.imshow(imageio.imread(url))
    plt.axis('off')
    plt.show()
    return pred

#Saving graphs
def save_viz(barchart,title):
    os.chdir('../Output')
    fig = barchart.get_figure()
    fig.savefig(title+ '.png')
    plt.show()
```

```python
if __name__ == '__main__':
    os.chdir('../Ironhack/Week 9/Final Project')
    artists = pd.read_csv('../Final Project/artists.csv')
    
    #graphs
    nationality=graph_nationality(artists)
    painters=graph_painters(artists)
    genre1=graph_genre1(artists)
    artists=genre_cleaning(artists)
    genre2=graph_genre2(artists)
    genre=['Baroque', 'Impressionism', 'Post-Impressionism', 'Renaissance', 'Symbolism']
    
    #creating X,y
    os.chdir( '../Final Project/images_resized_100_top_5' )
    X_list2=create_X()
    X=np.array(X_list2)
    index1=index1_def()
    y=create_y(index1)
    
    X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,stratify=y) # add stratify
    X_train = X_train / 255.0
    X_test = X_test / 255.0
    
    #showing images/genre
    original=original_genre()
    
    #transforming images
    transformed1,X_batch1,y_batch1=image_transformation1()
    transformed2,X_batch2,y_batch2=image_transformation2()
    transformed3,X_batch3,y_batch3=image_transformation3()
    
    X_train=X_train_transformed(X_train,X_batch1,X_batch2,X_batch3)
    y_train=y_train_transformed(y_train,y_batch1,y_batch2,y_batch3)
    
    #run model
    model=model(X_train,X_test,y_train,y_test)
    
    test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)
    print('\nTest accuracy:', test_acc)
    
    probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
    predictions = probability_model.predict(X_test)
    model.summary()
    
    # plotting some results
    results_graphs=plot_results(predictions)
    confusion_matrix=confusion_matrix_plot(X_test,y_test)
    
    #Saving output
    os.chdir( '../Output' )
    model.save('final_model_2.h5')
    
    #Trying on a new image from web
    url=input("please enter an image url to test the model !")
    web_image=transform_url(url)
    test_prediction=predict_url(web_image)
    
    #Saving graphs
    save_viz(nationality,"1. Nationalities of painters")
    save_viz(painters,"2. Names of painters")
    save_viz(genre1,"3. Genres of paintings")
    save_viz(genre2,"4. Cleaning genres of paintings")
    
    #Saving predictions and images
    save_viz(original,"5. Images and genres")    
    save_viz(results_graphs,"6. Model predictions")
    save_viz(confusion_matrix,"7. Confusion matrix")
    save_viz(test_prediction,"8. Test from web")
    
    #ending the program
    print("Thanks for running the program ! All graphs and the model are saved in the output folder")
```
